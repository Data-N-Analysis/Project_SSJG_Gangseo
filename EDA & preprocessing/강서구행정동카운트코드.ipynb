```python
def count_rows_with_keywords(dataframe, column_name, keywords):
    keyword_counts = {}
    for keyword in keywords:
        filtered_df = dataframe[dataframe[column_name].str.contains(keyword, na=False, regex=True)]
        keyword_count = len(filtered_df)
        keyword_counts[keyword] = keyword_count
    return keyword_counts
```
